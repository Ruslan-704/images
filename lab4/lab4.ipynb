```python
import sys
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('image.jpg')
plt.imshow(image)
```

```python
thresholds = [90, 160]
prev_image = image.copy()

for tr in thresholds:
    _, prev_image = cv2.threshold(prev_image, tr, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(14, 10))
plt.imshow(prev_image, cmap='gray')
plt.show()
```

```python
image = prev_image

kernel = np.ones((4, 4), np.uint8)
erosion = cv.erode(image, kernel, iterations=1)

plt.figure(figsize=(15, 8))
gs = plt.GridSpec(2, 3)

plt.figure(figsize=(14, 10))
plt.imshow(erosion, cmap='gray')
plt.show()

plt.show()

cv2.imwrite('processed_image.jpg', erosion)
```
